```python
import pandas as pd

ames_housing = pd.read_csv("house_prices.csv", na_values = "?")

target_name = "SalePrice"
data, target = (
    ames_housing.drop(columns = target_name),
    ames_housing[target_name],
)
target = (target > 200_000).astype(int)
```

```python
data
```

```python
# select features for simplicity

numeric_features = ["LotArea", "FullBath", "HalfBath"]
categorical_features = ["Neighborhood", "HouseStyle"]
data = data[numeric_features + categorical_features]
```

```python
# define preprocessing steps

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

numeric_transformer = Pipeline(
    steps = [
        ("imputer", SimpleImputer(strategy = "median")),
        ("scaler", StandardScaler(),),
    ]
)

categorical_transformer = OneHotEncoder(handle_unknown = "ignore")
```

```python
# apply transformations using ColumnTransformer

from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers = [
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)
```

```python
# define mode and join steps in order

from sklearn.linear_model import LogisticRegression

model = Pipeline(
    steps = [
        ("preprocessor", preprocessor),
        ("classifier", LogisticRegression()),
    ]
)
```

```python
#visualize pipeline

model
```

```python
# score model using cross validation

from sklearn.model_selection import cross_validate

cv_results = cross_validate(model, data, target, cv = 5)
scores = cv_results["test_score"]
print(
    "The mean cross-validation accuracy is: "
    f"{scores.mean():.3f} ± {scores.std():.3f}"
)
```
